## 6. Use Python to find the best-fit line. The Scikit-learn package is a good choice to use for this


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# read csv file
df = pd.read_csv(r"C:\Users\bharo\OneDrive\Documents\week2data (1)\week2data\car_data.csv")

x = df["weight"]
y = df["hwy_mpg"]

plt.scatter(x,y)
plt.title("Miles per Gallon VS Weight", color = "Green")
plt.xlabel("Weight", color = "Green")
plt.ylabel("Highway miles per gallon", color = "Green") 

#convert x dataframe to two dimentional array
x1 = np.array(x).reshape(-1,1)

#importing LinearRegression from Sciket-learn package to plot the line
from sklearn.linear_model import LinearRegression

# Assigning the funtion to LR variable
LR = LinearRegression()
#fitting x1 and y to LR
LR.fit(x1,y)

#LR.predict() calcultes mx+b 
y1 = LR.predict(x1) 
c = LR.intercept_
m1 = LR.coef_

plt.plot(x1,y1,color = "Green")
print( c, m1 )
```
